```python
import pygame
import random

pygame.init()

WIDTH, HEIGHT = 900, 500
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Space Miner Game")

font = pygame.font.SysFont(None, 28)

# Game variables
fuel = "High"
health = "Good"
score = 0
turn = 0
max_turns = 10
message = "Press 1 Safe | 2 Aggressive | 3 Refuel"

running = True


def draw_screen():

    screen.fill((20,20,30))

    # ---- Ship Area ----
    pygame.draw.rect(screen,(40,40,80),(0,0,300,300),2)
    ship_text = font.render("Ship Area",True,(255,255,255))
    screen.blit(ship_text,(110,10))

    pygame.draw.rect(screen,(0,200,255),(130,120,40,40))  # Ship


    # ---- Asteroid Field ----
    pygame.draw.rect(screen,(40,40,80),(300,0,300,300),2)
    asteroid_text = font.render("Asteroid Field",True,(255,255,255))
    screen.blit(asteroid_text,(370,10))

    for i in range(5):
        pygame.draw.circle(screen,(150,150,150),(350+i*50,150),15)


    # ---- Score Panel ----
    pygame.draw.rect(screen,(40,40,80),(600,0,300,300),2)

    score_text = font.render(f"Score: {score}",True,(255,255,255))
    screen.blit(score_text,(680,50))

    health_text = font.render(f"Health: {health}",True,(255,255,255))
    screen.blit(health_text,(680,100))

    turn_text = font.render(f"Turns: {turn}/{max_turns}",True,(255,255,255))
    screen.blit(turn_text,(680,150))


    # ---- Fuel Bar ----
    fuel_label = font.render("Fuel:",True,(255,255,255))
    screen.blit(fuel_label,(40,320))

    if fuel == "High":
        pygame.draw.rect(screen,(0,255,0),(100,320,120,20))
    else:
        pygame.draw.rect(screen,(255,0,0),(100,320,60,20))


    # ---- Controls ----
    controls = font.render("1 Safe Mine   2 Aggressive Mine   3 Refuel",True,(255,255,0))
    screen.blit(controls,(220,420))

    msg = font.render(message,True,(255,255,255))
    screen.blit(msg,(250,380))

    # ---- Game Over ----
    if turn >= max_turns:
        game_over = font.render(f"GAME OVER! Final Score: {score}",True,(255,80,80))
        screen.blit(game_over,(300,250))

    pygame.display.update()


while running:

    draw_screen()

    for event in pygame.event.get():

        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN and turn < max_turns:

            if event.key == pygame.K_1:  # SAFE
                turn += 1
                r = random.random()

                if r < 0.8:
                    score += 5
                    message = "Safe mining success (+5)"
                else:
                    message = "Nothing found"

                fuel = "Low"


            elif event.key == pygame.K_2:  # AGGRESSIVE
                turn += 1
                r = random.random()

                if r < 0.5:
                    score += 12
                    message = "Big minerals (+12)"

                elif r < 0.8:
                    score += 5
                    message = "Small minerals (+5)"

                else:
                    score -= 8
                    health = "Damaged"
                    message = "Ship damaged (-8)"

                fuel = "Low"


            elif event.key == pygame.K_3:  # REFUEL
                turn += 1
                fuel = "High"
                score -= 1
                message = "Refueled (-1)"


pygame.quit()
```
